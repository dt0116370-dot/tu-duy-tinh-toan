Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def is_armstrong(n: int) -> bool:
    if n<0:
        return False
    s=str(n)
    k=len(s)
    total=0
    for i in s:
        total += int(i)**k
    return total == n

n=int(input())
print(is_armstrong(n))
```

Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def check(x: int) -> bool:
    for i in range(1,x+1):
        if i*i==x:
            return True
    return False

x=int(input())
print(check(x))
```

W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'

```python
def detect_number_type(s) :
    try:
         int(s)
         return 'int'
    except ValueError:
        try :
            float(s)
            return 'float'
        except ValueError:
            return 'neither'
    
s=input()
print(detect_number_type(s))
```

W5A23 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0

```python
def quadratic_roots_count(a, b, c):
    if a==0:
        x=(-c)/b
        return x
    elif a != 0:
        denta= (b**2) - 4*a*c
        if denta < 0 :
            return 'Vo nghiem'
        elif denta == 0:
            x1=x2=(-b)/(2*a)
            return x1==x2
        elif denta > 0:
            x1=((-b)+(denta**0.5))/2*a
            x2=((-b)-(denta**0.5))/2*a
            return x1,x2

a,b,c=map(int,input().split())
print(quadratic_roots_count(a, b, c))
```

W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không

```python
def are_collinear(x1,y1,x2,y2,x3,y3):
    S= x1 * (y2-y3) + x2 * (y3-y1) + x3 * (y1-y2)
    if S == 0 :
      return True
    else:
       return False
    
x1,y1,x2,y2,x3,y3=map(int,input().split())
print(are_collinear(x1,y1,x2,y2,x3,y3))
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3)

```python
import math
def sdg(x1, y1, x2, y2, x3, y3):
    BAx,BAy=x2-x1,y2-y1
    BCx,BCy=x3-x2,y3-y2
    tvh=(x2-x1)*(x3-x2)+(y2-y1)*(y3-y2)
    BA=((x2-x1)**2+(y2-y1)**2)**0.5
    BC=((x3-x2)**2+(y3-y2)**2)**0.5
    if BA == 0 or BC == 0:
        return 0
    cos=tvh/(BA*BC)
    cos=max(-1,min(1,cos))
    deg=math.degrees(cos)
    return round(deg,2)

x1, y1, x2, y2, x3, y3=map(int,input().split())
print(sdg(x1, y1, x2, y2, x3, y3))
```

Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa,
giá km, 

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    if car_type == 7:
        open_door= 20000
        duoi_25km= 15000
        tren_25km= 13000
    else:
        open_door = 15000
        duoi_25km = 13000
        tren_25km = 10000

    tong=0
    if km <= 0.5:
        tong = open_door
    elif km <= 25:
        tong = open_door + (km-0.5)*duoi_25km
    else:
        tong= open_door + 24.5*duoi_25km +(km-25)*tren_25km

    if wait_minutes > 5:
        tong += (wait_minutes-5)*1000
    if start_hour >= 22 and start_hour < 5:
        tong *= 1.1
    return f"{tong:.0f}"

car_type, start_hour, km, wait_minutes=map(int,input().split())
print(smart_taxi_fare(car_type, start_hour, km, wait_minutes))

```

Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau


```python
def check(x):
    s=str(x)
    return  len(s)==len(set(s))
def sum_1_to_n(n):
    tong=0
    for x in range(1,n+1):
        if check(x):
            tong += x
    return tong

n=int(input())
print(sum_1_to_n(n))
```
