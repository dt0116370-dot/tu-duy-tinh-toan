Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình

```python
a,b,c=map(float,input().split())
if a+b-c>0 and a+c-b>0 and a+c-b>0:
    t=(a+b+c)/2
    c=a+b+c
    s=(t*(t-a)*(t-b)*(t-c))**0.5
    print("diện tích tam giác là: ",f"{s:.0f}")
    print("chu vi tam giác là: ",f"{c:.0f}")
    if a == b and b == c :
        print("tam giác đều")
    elif a==b or a==c or b == c :
        print("tam giác cân")
    elif (a**2==b**2+c**2) or (b**2==a**2+c**2) or (c**2==a**2+a**2):
        print("tam giác vuông")
    else:
        print("tam giac thuong")
else:
    print("khong la tam giac")
```
