W5A2 Viết hàm hoán đổi hai số nguyên. def swap(a, b) -> Tuple[int, int]: 


```python
def swap(a,b):
    a,b=b,a
    return a,b
a,b=swap(a,b)
a,b=input().split()
print(swap(a,b))
```


W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 


```python
def check(n):
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True

n=int(input())
print(check(n))
```

W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 

```python
def check(n):
    tong=0
    for i in range(1,n):
        if n%i==0:
            tong += i
    if tong == n:
        return True
    else:
        return False

n=int(input())
print(check(n))
```

viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử
đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1 


```python
def check(a,k):
    d=0
    for i in a:
        d+=1
        if i == k:
            return d
    return -1

a=list(map(int,input().split()))
k=int(input())
print(check(a,k))
```

W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)

```python
def giai_thua(n):
    tich=1
    for i in range(1,n+1):
        tich *= i
    return tich

n=int(input())
print(giai_thua(n))
```

Viết hàm giả lập một máy tính bỏ túi.
Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat
\)
Biết các toán tử đơn giản thực hiện phép toán lần lượt là:
Phép trừ \( - \)
Phép cộng \( + \)
Phép chia \( / \)
Phép nhân \( * \)
Kết quả làm tròn đến chữ số thập phân thứ hai.
Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)).
Đầu ra: Kết quả phép toán

```python
def casio(num1:float,num2:float,dau:str):
    if dau == '+':
        kqua= num1 + num2
    elif dau == '-':
        kqua = num1 - num2
    elif dau == '*':
        kqua = num1 * num2
    elif dau == '/':
        if num2 == 0:
            return False
        else:
            kqua=num1/num2
    return kqua
 
print(casio(num1:float,num2:float,dau:str))
```

W5A15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông

```python
def is_right_triangle(a,b,c):
    if b**2 == a**2 + c**2 or a**2 == b**2 + c**2 or c**2 == a**2 + b**2:
        return True
    else:
        return False  
    
a=int(input())
b=int(input())
c=int(input())
print(is_right_triangle(a,b,c))
```

W5A11 Viết hàm is_product_divisible_by_10(a, b) trả về True nếu tích a*b chia hết cho 10, ngược lại False

```python
def check(a,b):
    x=a*b
    if x%10==0:
        return True
    else:
        return False
    
a,b=map(int,input().split())
print(check(a,b))
```

Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op. 


```python

def compute(a):
    b=float(a[0])
    c=float(a[1])
    op=str(a[2])
    if op == '+':
        return b + c
    elif op == '-':
        return b - c
    elif op == '*':
        return b * c
    elif op == '/':
        return f"{(b/c):.2f}"
a=list(map(str,input().split()))    
print(compute(a))
```
